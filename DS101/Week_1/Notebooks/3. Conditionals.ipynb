## Everything is a Bool.

<img src="http://www.trytoprogram.com/images/python_bool_function.jpg">

#### `type(bool)`

```python
print('Type of "True": {}'.format(type(True)))
print('Type of "False": {}'.format(type(False)))```

#### `bool(number)`

```python
print('0: {}'.format(bool(0)))
print('1: {}'.format(bool(1)))```

#### `bool(list)`

```python
print('Empty list: {}'.format(bool([])))
print('list with values: {}'.format(bool(["a","b"])))```

#### `bool(dict)`

```python
print('Empty dict: {}'.format(bool({})))
print('dict with values: {}'.format(bool({'Python': 'cool'})))```

***
<br>
<br>

## Conditional Operators

#### Equal To `==`

```python
print('1 == 0: {}'.format(1 == 0))```

#### Not-Equal To `!=`

```python
print('1 != 0: {}'.format(1 != 0))```

#### Greater Than `>`

```python
print('1 > 0: {}'.format(1 > 0))
print('1 > 1: {}'.format(1 > 1))```

#### Less Than `<`

```python
print('1 < 0: {}'.format(1 < 0))
print('1 < 1: {}'.format(1 < 1))```

#### Greater Than or Equal To `>=` 

```python
print('1 >= 0: {}'.format(1 >= 0))
print('1 >= 1: {}'.format(1 >= 1))```

#### Less Than or Equal To `<=`

```python
print('1 <= 0: {}'.format(1 <= 0))
print('1 <= 1: {}'.format(1 <= 1))```

#### We can combine them!

```python
bool_operation = 0 <= 1 <= 2
print('1 <= 2 <= 3: {}'.format(bool_operation))```

***
<br>
<br>

## Logical Operators

### `and`

```python
print('True and True: {}'.format(True and True))
print('True and False: {}'.format(True and False))```

### `or`

```python
print('True or True: {}'.format(True or True))
print('True or False: {}'.format(True or False))```

### `not`

```python
python_is_cool = False
print('Python is cool: {}'.format(not python_is_cool))```

### `in`

```python
test = 'd' in 'dog'
print('is "d" in "dog"?', test)```

#### We can combine them! 
>Be sure to use parentheses to group operations!

```python
eval_false = True and False
eval_true = True or False
print("(True and False) or (True or False): {}".format(eval_false or eval_true))```

***
<br>
<br>

## If - Statements

<img width = "500px"src="https://data36.com/wp-content/uploads/2018/01/1-Python-if-statement-logic-768x374.png">

#### Basic Example

```python
statement = True

if statement:
    print("statement is True")
    
if not statement:
    print("statement is not True")```

#### With `if()` statements, conversion to `bool` is implicit.

```python
empty_list = []

if empty_list:
    print('empty list will not evaluate to True')  # this won't be executed```

#### `if()` statements can handle *multiple* conditions.

```python
val = 3

if (0 <= val < 1 ) or (val == 3):
    print('Value is positive and less than one --or-- value is three')```

<br>

### `if-else`

```python
my_dict = {}

if my_dict:
    print("there is something in my dict")
    
else:
    print("my dict is empty")```

<br>

### `if-elif-else`

```python
val = 110

if val >= 100:
    print("value is equal or greater than 100")
    
elif val > 10:
    print("value is greater than 10 but less than 100")
    
else:
    print("value is equal or less than 10")```

<br>

#### You can have as many `elif` statements as you need. In addition, `else` at the end is not mandatory.

```python
# Hint: Change value of language

language = 'Swedish'

if language == 'Swedish':
    special_greeting = 'Hejsan!'
elif language == 'Finnish':
    special_greeting = 'Latua perkele!'
elif language == 'Spanish':
    special_greeting = 'Hola!'
elif language == 'German':
    special_greeting = 'Guten Tag!'
else:
    special_greeting = 'Hello!'

greeting = '{} fellow Pythonista!'.format(special_greeting)
print(greeting)```

For more detailed overview about conditionals, check this [tutorial from Real Python](https://realpython.com/python-conditional-statements/).
